```python
#EJERCICIO 1 MEDIA; MIN Y MAX:
def media_edad(m):
    if not m:
        return 0.0

    total_edades = 0
    total_usuarios = 0

    for usuario in m:
        total_edades += usuario[2]
        total_usuarios += 1

    return total_edades / total_usuarios

def media_de_un_usuario(usuario):
    if len(usuario) < 4:
        return 0.0

    total_compras = sum(usuario[3:])
    numero_compras = len(usuario) - 3

    if numero_compras == 0:
        return 0.0

    return total_compras / numero_compras

def max_de_medias(m):
    if not m:
        return None

    max_media = 0
    max_usuario = ""

    for usuario in m:
        media = media_de_un_usuario(usuario)
        if media > max_media:
            max_media = media
            max_usuario = usuario[1]

    return max_usuario

def rango_de_edad(m):
    if not m:
        return (0, 0)

    min_edad = float('inf')
    max_edad = float('-inf')

    for usuario in m:
        edad = usuario[2]
        if edad < min_edad:
            min_edad = edad
        if edad > max_edad:
            max_edad = edad

    return (min_edad, max_edad)

usuarios = [['1234567D','Joan',50, 210, 20, 40],
            ['4324D','Antonio',69,710,550,400,698,645,512],
            ['5544C','Claudio',72,100,534],
            ['5432F','Maria',75,770,645,230,650,590,481,602]]

# Pruebas
assert media_edad(usuarios) == 66.5
assert media_de_un_usuario(['1234567D','Joan',50, 210, 20, 40]) == 90.0
assert max_de_medias(usuarios) == "Antonio"
assert rango_de_edad(usuarios) == (50, 75)

#EJERCICIO 2 NOTAS
def aprobados(m):
    aprobados = []
    for alumno in m:
        notas = alumno[1:]
        if all(nota >= 5.0 for nota in notas):
            aprobados.append(alumno[0])
    return aprobados

def alumno_mejor_nota(m):
    mejor_nota = -1.0
    id_alumno = ""
    for alumno in m:
        notas = alumno[1:]
        max_nota = max(notas)
        if max_nota > mejor_nota:
            mejor_nota = max_nota
            id_alumno = alumno[0]
    return id_alumno

def alumno_mejor_nota_media(m):
    mejor_media = -1.0
    id_alumno = ""
    for alumno in m:
        notas = alumno[1:]
        media = sum(notas) / len(notas)
        if media > mejor_media:
            mejor_media = media
            id_alumno = alumno[0]
    return id_alumno

def alumnos_peores_notas(m, n):

    m.sort(key=lambda x: sum(x[1:]) / len(x[1:]))


    peores_notas = [alumno[0] for alumno in m[:n]]
    return peores_notas


list_notas = [['33333R', 0.0, 9.0, 8.5, 6.0],
              ['337464F', 8.5, 7.0, 2.0, 8.5],
              ['00000F', 0.5, 1.0, 2.0, 8.5],
              ['112233A', 5.0, 6.0, 8.5, 4.5]]

assert aprobados([['40970455X', 5, 4, 6], ['896737498N', 7, 7, 0]]) == ['40970455X']
assert aprobados(list_notas) == ['33333R', '337464F', '112233A']

assert alumno_mejor_nota(list_notas) == "33333R"  # Mejor nota 9.0

assert alumno_mejor_nota_media([['40970455X', 5, 4, 10], ['896737498N', 7, 7, 10]]) == "896737498N"
assert alumno_mejor_nota_media(list_notas) == "337464F"

assert alumnos_peores_notas(list_notas, 2) == ['00000F', '33333R']

#EJERCICIO 3 COMPROBAR ERRORES DE ECUACION

def comprobar_parentesis_con_texto(texto):
    pila = []
    mapeo = {')': '(', ']': '[', '}': '{'}

    for char in texto:
        if char in '([{':
            pila.append(char)
        elif char in ')]}':
            if not pila or pila[-1] != mapeo[char]:
                return False
            pila.pop()

    return len(pila) == 0


assert comprobar_parentesis_con_texto("(-5[{2 + x}])")
assert not comprobar_parentesis_con_texto("(3 - 29[2{40+2(}])")
assert comprobar_parentesis_con_texto("(x and y or ( not x and [ z or {z and y}]))")
assert comprobar_parentesis_con_texto("-3{2x + 5}-(y[xy{3z+4}])")
assert not comprobar_parentesis_con_texto("-3}2x + 5{-(y[xy{3z+4}])")
assert not comprobar_parentesis_con_texto("{")


#EJERCICIO 4 simulador_ejercicio_prioridad
from queue import PriorityQueue

def anadir_paquete(servidor, caducidad, prioridad):
    servidor.put((caducidad, prioridad))

def procesar(servidor, tiempo):
    paquetes_descartados = 0
    while not servidor.empty():
        caducidad, prioridad = servidor.get()
        if caducidad <= tiempo:
            paquetes_descartados += 1
        else:
            return paquetes_descartados
    return paquetes_descartados

def simulador_servidor_prioridad(path_in, path_out):
    servidor = PriorityQueue()
    paquetes_perdidos = 0
    tiempo = 0
    with open(path_in) as f, open(path_out, "w") as f_out:
        lineas = f.readlines()
        for linea in lineas:
            partes = linea.split()
            comando = partes[0]
            if comando == 't':
                tiempo = int(partes[1])
            elif comando == 'paquete':
                caducidad = int(partes[1])
                prioridad = int(partes[2])
                anadir_paquete(servidor, caducidad, prioridad)
            elif comando == 'perdidos':
                paquetes_perdidos = procesar(servidor, tiempo)
                f_out.write(f'{paquetes_perdidos}\n')


simulador_servidor_prioridad("test_server_prioridad.txt", "test_server_prioridad_salida.txt")
```
