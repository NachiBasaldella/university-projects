```python
#ACTIVIDAD 5
      #Número de digitos
def num_digitos(n):

    if n < 10:
        return 1

      else:
        return 1 + num_digitos(n // 10)

assert(num_digitos(0) == 1)
assert(num_digitos(3) == 1)
assert(num_digitos(10) == 2)
assert(num_digitos(23) == 2)
assert(num_digitos(123) == 3)
assert(num_digitos(1001111) == 7)

        #Binario
# Función utilizando strings
def binario_str(n):
    if n == 0:
        return '0'
    elif n == 1:
        return '1'
    else:
        #obtener el conciente
        cociente = n // 2
        # obtener el residuo y convertirlo a string
        residuo = str(n % 2)
        # Combinar el residuo y el resultado
        return binario_str(cociente) + residuo

# Función utilizando enteros
def binario_int(n):
    if n == 0:
        return 0
    elif n == 1:
        return 1
    else:
        #obtener el cociente
        cociente = n // 2
        #obtener el residuo y combinarlo con el resultado
        return binario_int(cociente) * 10 + n % 2


assert (binario_str(0)    == '0')
assert (binario_str(1)    == '1')
assert (binario_str(2)    == '10')
assert (binario_str(5)    == '101')
assert (binario_str(1024) == '10000000000')
assert (binario_str(200)  == '11001000')

assert (binario_int(0)    == 0)
assert (binario_int(1)    == 1)
assert (binario_int(2)    == 10)
assert (binario_int(5)    == 101)
assert (binario_int(1024) == 10000000000)
assert (binario_int(200)  == 11001000)

        #Doble factorial de un número
def doble_factorial(n):
    if n <= 1:
        return 1
    else:
        return n * doble_factorial(n - 2)


assert doble_factorial(1) == 1
assert doble_factorial(2) == 2
assert doble_factorial(3) == 3
assert doble_factorial(4) == 8
assert doble_factorial(8) == 384
assert doble_factorial(9) == 945


      #Raíz de la ecuación
def biseccion(tol, a, b, f):
       c = (a + b) / 2

    fa = f(a)
    fb = f(b)
    fc = f(c)


    if abs(fc) < tol or (b - a) / 2 < tol:
        return c


    if fa * fc < 0:
        return biseccion(tol, a, c, f)
    else:
        return biseccion(tol, c, b, f)

# Función de prueba 1
def f(x):
    return x - 3

a = 0
b = 5
p = 0.01

resultado1 = biseccion(p, a, b, f)
print(resultado1)

# Función de prueba 2
def g(x):
    return x**2 - 25

a = 0
b = 22
p = 0.1

resultado2 = biseccion(p, a, b, g)
print(resultado2)
```
